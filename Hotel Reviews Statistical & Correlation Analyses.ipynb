```python
import pandas as pd

# Load the dataset
df = pd.read_csv(r'C:\Users\yashe\Downloads\Hotel_Reviews.csv')

# Display basic information about the dataset
print(df.info())
df.head()
```

```python
# Select relevant columns for analysis
relevant_columns = [
    "Reviewer_Score", 
    "Review_Total_Negative_Word_Counts", 
    "Review_Total_Positive_Word_Counts", 
    "Additional_Number_of_Scoring", 
    "Total_Number_of_Reviews", 
    "Total_Number_of_Reviews_Reviewer_Has_Given",
    "Average_Score",
    "lat",
    "lng"
]

df_relevant = df[relevant_columns]

# Display the first few rows of the filtered dataset
print(df_relevant.info())
df_relevant.head()
```

```python
# Check for missing values
print(df_relevant.isnull().sum())

# Drop rows with missing values
df_relevant = df_relevant.dropna()

# Confirm there are no missing values left
print(df_relevant.isnull().sum())
```

```python
# Calculate the correlation matrix
correlation_matrix = df_relevant.corr()

# Display correlation values for Reviewer_Score
print(correlation_matrix["Reviewer_Score"].sort_values(ascending=False))

# Plot the correlation heatmap
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Matrix")
plt.show()
```

```python
# Scatter plot: Reviewer_Score vs Review_Total_Positive_Word_Counts
plt.figure(figsize=(8, 5))
sns.scatterplot(data=df_relevant, x="Review_Total_Positive_Word_Counts", y="Reviewer_Score")
plt.title("Reviewer Score vs Positive Word Counts")
plt.xlabel("Positive Word Counts")
plt.ylabel("Reviewer Score")
plt.show()

# Scatter plot: Reviewer_Score vs Average_Score
plt.figure(figsize=(8, 5))
sns.scatterplot(data=df_relevant, x="Average_Score", y="Reviewer_Score", color='red')
plt.title("Reviewer Score vs Average Score")
plt.xlabel("Average Score")
plt.ylabel("Reviewer Score")
plt.show()

# Scatter plot: Reviewer_Score vs Review_Total_Negative_Word_Counts
plt.figure(figsize=(8, 5))
sns.scatterplot(data=df_relevant, x="Review_Total_Negative_Word_Counts", y="Reviewer_Score", color='green')
plt.title("Reviewer Score vs Negative Word Counts")
plt.xlabel("Negative Word Counts")
plt.ylabel("Reviewer Score")
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
import numpy as np

# Linear regression for positive word counts
X_positive = df_relevant[["Review_Total_Positive_Word_Counts"]]
y = df_relevant["Reviewer_Score"]

model_positive = LinearRegression()
model_positive.fit(X_positive, y)

print(f"Positive Words Regression Coefficient: {model_positive.coef_[0]}")
print(f"Intercept: {model_positive.intercept_}")

# Linear regression for negative word counts
X_negative = df_relevant[["Review_Total_Negative_Word_Counts"]]

model_negative = LinearRegression()
model_negative.fit(X_negative, y)

print(f"Negative Words Regression Coefficient: {model_negative.coef_[0]}")
print(f"Intercept: {model_negative.intercept_}")

# Linear regression for average score
X_negative = df_relevant[["Average_Score"]]

model_negative = LinearRegression()
model_negative.fit(X_negative, y)

print(f"Average Score Regression Coefficient: {model_negative.coef_[0]}")
print(f"Intercept: {model_negative.intercept_}")
```

```python
import statsmodels.api as sm

X = df_relevant["Review_Total_Positive_Word_Counts"]
y = df_relevant["Reviewer_Score"]
X = sm.add_constant(X)  # Add constant for intercept

model = sm.OLS(y, X).fit()
print(model.summary())

# Plot residuals
plt.figure()
sm.qqplot(model.resid, line="45")
plt.title("Residuals QQ Plot")
plt.show()
```

```python
# Define a function to fit the model and return the R-squared value
def get_r_squared(X, y):
    X = sm.add_constant(X)
    model = sm.OLS(y, X).fit()
    return model.rsquared

#Additional_Number_of_Scoring vs Reviewer_Score
X = df_relevant["Additional_Number_of_Scoring"]
y = df_relevant["Reviewer_Score"]
r_squared_1 = get_r_squared(X, y)
print(f"R-squared for Additional_Number_of_Scoring: {r_squared_1}")

#Total_Number_of_Reviews vs Reviewer_Score
X = df_relevant["Total_Number_of_Reviews"]
y = df_relevant["Reviewer_Score"]
r_squared_2 = get_r_squared(X, y)
print(f"R-squared for Total_Number_of_Reviews: {r_squared_2}")

#Total_Number_of_Reviews_Reviewer_Has_Given vs Reviewer_Score
X = df_relevant["Total_Number_of_Reviews_Reviewer_Has_Given"]
y = df_relevant["Reviewer_Score"]
r_squared_3 = get_r_squared(X, y)
print(f"R-squared for Total_Number_of_Reviews_Reviewer_Has_Given: {r_squared_3}")

#lat vs Reviewer_Score
X = df_relevant["lat"]
y = df_relevant["Reviewer_Score"]
r_squared_4 = get_r_squared(X, y)
print(f"R-squared for lat: {r_squared_4}")

#lng vs Reviewer_Score
X = df_relevant["lng"]
y = df_relevant["Reviewer_Score"]
r_squared_5 = get_r_squared(X, y)
print(f"R-squared for lng: {r_squared_5}")
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

# Function to calculate VIF for each variable
def calculate_vif(X):
    vif_data = pd.DataFrame()
    vif_data["Variable"] = X.columns
    vif_data["VIF"] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
    return vif_data

# Exclude the target variable (Reviewer_Score) and include only independent variables
X_vif = df_relevant[[
    "Review_Total_Negative_Word_Counts", 
    "Review_Total_Positive_Word_Counts", 
    "Average_Score"
]]

# Add a constant for VIF calculation
X_vif = sm.add_constant(X_vif)

# Calculate and display VIF
vif_data = calculate_vif(X_vif)
print(vif_data)
```
